```python
import pandas as pd
```

```python
# Loading the CSV file into the DataFrame
df = pd.read_csv('Financial_Report.csv')
```

```python
df
```

```python
#Fill the NaN values in Columns 'Total Liabilities','Cash Flow from Operating Activities' with Mean
df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']] = df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']].fillna(df[['Total Assets','Total Liabilities','Cash Flow from Operating Activities']].mean())
```

```python
df
```

Now calculating year-over-year changes for each financial metric. You can do this by creating new columns in your DataFrame that represent the percentage change from one year to the next.

```python
# calculating year-over-year changes for each financial metric
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Total Assets Change (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Total Liabilities Change (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Operating Activities Change (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

Explore other aggregate functions or groupings to analyze the data across different dimensions (by company, over years, etc.)

```python
# Groupings by company
by_company = df.groupby('Company').agg({
    'Total Revenue': ['mean', 'median', 'min', 'max'],
    'Net Income': ['mean', 'median', 'min', 'max'],
    'Total Assets': ['mean', 'median', 'min', 'max'],
    'Total Liabilities': ['mean', 'median', 'min', 'max'],
    'Cash Flow from Operating Activities': ['mean', 'median', 'min', 'max']
})

# Groupings over years
over_years = df.groupby(df.Year).agg({
    'Total Revenue': ['sum', 'mean', 'median', 'min', 'max'],
    'Net Income': ['sum', 'mean', 'median', 'min', 'max'],
    'Total Assets': ['sum', 'mean', 'median', 'min', 'max'],
    'Total Liabilities': ['sum', 'mean', 'median', 'min', 'max'],
    'Cash Flow from Operating Activities': ['sum', 'mean', 'median', 'min', 'max']
})

# Groupings by company and over years
by_company_and_years = df.groupby(['Company', df.Year]).agg({
    'Total Revenue': 'sum',
    'Net Income': 'sum',
    'Total Assets': 'sum',
    'Total Liabilities': 'sum',
    'Cash Flow from Operating Activities': 'sum'
})
```

```python
#Displaying the results
print("Aggregate functions/groupings by company:")
print(by_company)
print("\nAggregate functions/groupings over years:")
print(over_years)
print("\nAggregate functions by company and over years:")
print(by_company_and_years)
```
